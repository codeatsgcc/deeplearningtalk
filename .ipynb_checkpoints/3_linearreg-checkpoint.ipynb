### Introduction

In this notebook, we show how to run a linear regression using the `scikit-learn` package as demonstrated during the talk

We will first generate a "fake" dataset of the height(cm) and weight(kg) of 1000 people by simulation with the following assumptions:
- height and weight are Normally distributed $N(165,10)$ and $N(55,7)$ respectively

We will then build a linear regression model to predict weight using height as covered during the talk

### Step 1: Simulating the data

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
sns.set()
```

```python
mu_x = 165
sig2_x = 10
sig_x = np.sqrt(sig2_x)

mu_y = 55
sig2_y = 7
sig_y = np.sqrt(sig2_y)

corr = 0
cov = corr * sig_x * sig_y

MU = (mu_x, mu_y)
SIGMA = (
(sig2_x, cov),
(cov  , sig2_y)
)

x = np.random.multivariate_normal(MU, SIGMA, 1000)
```

```python
students = pd.DataFrame(x, columns=('height','weight'))
```

```python
students.values[:10,:]
```

```python
plt.scatter(students['height'], students['weight'] )
plt.xlabel("height (cm)")
plt.ylabel("weight (kg)")
```

```python
sns.jointplot(students['height'], students['weight'] )
```

### Step 2: Running a Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(students['height'].values[:,np.newaxis], students['weight'].values[:,np.newaxis])

max_height = np.round(np.max(students['height']))+1
min_height = np.round(np.min(students['height']))-1
xfit = np.linspace(min_height, max_height, 1000)
yfit = model.predict(xfit[:,np.newaxis])
plt.scatter(students['height'], students['weight'])
plt.plot(xfit,yfit)
```
